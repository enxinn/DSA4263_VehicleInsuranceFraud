```python
import nbformat
from nbclient import NotebookClient
from multiprocessing import Pool
import time
import os
```

```python
def run_notebook(path):
    start = time.time()
    print(f"\n=== Running {path} ===")

    nb = nbformat.read(path, as_version=4)
    client = NotebookClient(nb, timeout=10000, kernel_name="python3", allow_errors=False)
    client.execute()
    nbformat.write(nb, path)

    print(f"Finished {path} in {(time.time()-start)/60:.2f} min")
    return path
```

```python
# Run all notebooks in sequence
notebooks = [
    "notebooks/01_eda_preprocessing.ipynb",
    "notebooks/02_resampling_method.ipynb",
    "notebooks/03_logistic_regression.ipynb",
    "notebooks/04_random_forest.ipynb",
    "notebooks/05_decision_tree.ipynb",
    "notebooks/06_lightgbm.ipynb",
    "notebooks/07_xgboost.ipynb",
    "notebooks/08_svc.ipynb",
    "notebooks/09_lightgbm_finalmodel.ipynb"
]

for nb in notebooks:
    run_notebook(nb)
```
